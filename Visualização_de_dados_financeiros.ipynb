# Análise e visualização de dados de ações

Neste estudo vamos extrair informações de ações do banco de dados do Yahoo Finance usando o pacote 'yfinance'.

Todas informações sobre a biblioteca no site: https://pypi.org/project/yfinance/

# Importando Bibliotecas


```python
# Instalação da biblioteca do yahoo finance
!pip install -q yfinance
```

```python
# Importação
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

# Base de dados de ações

```python
# Criando lista de ações que seram analisadas
acoes = ['ABEV3.SA', 'ODPV3.SA', 'VIVT3.SA', 'PETR3.SA', 'BBAS3.SA', 'BOVA11.SA']
```

```python
# Visualizando a lista
acoes
```

```python
# Buscando os valores de cada ação
acoes_df = pd.DataFrame()
for acao in acoes:
  acoes_df[acao] = yf.download(acao, start='2015-01-01', end='2020-11-04')['Close']
```

```python
# Visualizando data frame
acoes_df
```

```python
# Alterando nomes de colunas
acoes_df = acoes_df.rename(columns={'ABEV3.SA':'AMBEV', 'ODPV3.SA':'ODONTOPREV', 'VIVT3.SA':'VIVO',
                                    'PETR3.SA':'PETROBRAS', 'BBAS3.SA':'BBRASIL', 'BOVA11.SA':'BOVA'})
```

```python
# Visualizando as alterações
acoes_df
```

```python
# Salvando dataset
acoes_df.to_csv('acoes_ex.csv')
```

```python
# Carregando dataset
df = pd.read_csv('acoes_ex.csv')
df
```

```python
# Visualizando as informações do dataset
df.info()
```

```python
# Estatisticas do dataset
df.describe()
```

```python
# Visualizando histograma de cada ação
plt.figure(figsize=(10,50))
i = 1
for i in np.arange(1, len(df.columns)):
  plt.subplot(7, 1, i + 1)
  sns.histplot(df[df.columns[i]], kde=True)
  plt.title(df.columns[i]);
```

```python
# Visualizando boxplot de cada ação
plt.figure(figsize=(10,50))
i = 1
for i in np.arange(1, len(df.columns)):
  plt.subplot(7, 1, i + 1)
  sns.boxplot(x=df[df.columns[i]])
  plt.title(df.columns[i]);
```

```python
# Calculando o crescimento proporcional dos preços das ações
df_normalizado = df.copy()
for i in df_normalizado.columns[1:]:
  df_normalizado[i] = df_normalizado[i] / df_normalizado[i][0]
```

```python
# Visualização
df_normalizado
```

```python
# Visualização das ações em gráfico dinâmico
figura = px.line(title='Histórico do preço das ações')
for i in df.columns[1:]:
  figura.add_scatter(x=df['Date'], y=df[i], name=i)
figura.show()
```
